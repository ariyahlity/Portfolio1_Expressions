```python
# BMI Calculator
# The program calculates the BMI of the user by the use of variables, arithmetic operations, and expressions 

# prints the BMI CALCULATOR
print("----------------- BMI CALCULATOR -----------------")

# prompts the user to enter the weight in kilograms 
weight = float(input("Enter your weight in kg: "))

# prompts the user to enter the height in meters
height = float(input("Enter your height in m: "))

# calculates the BMI of the user by dividing the weight over the height squared
BMI = weight / (height * height)    

# prints the output
print(f"\nYour BMI is {BMI}")
```
